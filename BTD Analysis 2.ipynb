```python
def percent_missing(bnk_c):
    percent_nan = 100* bnk_c.isnull().sum() / len(bnk_c)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(bnk_c)
```
